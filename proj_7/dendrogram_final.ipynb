```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
from sklearn import datasets
from sklearn.metrics import confusion_matrix
```

```python
iris = datasets.load_iris()
```

```python
plt.figure(figsize=(25, 10))
plt.title('Iris Hierarchical Clustering Dendrogram')
plt.xlabel('Species')
dendrogram = sch.dendrogram(sch.linkage(iris.data[:,[0,1]], method='single'))
```

```python
plt.figure(figsize=(25, 10))
plt.title('Iris Hierarchical Clustering Dendrogram')
plt.xlabel('Species')
dendrogram = sch.dendrogram(sch.linkage(iris.data[:,[0,1]], method='complete'))
```

```python
plt.figure(figsize=(25, 10))
plt.title('Iris Hierarchical Clustering Dendrogram')
plt.xlabel('Species')
dendrogram = sch.dendrogram(sch.linkage(iris.data[:,[0,1]], method='average'))
```

```python
plt.figure(figsize=(25, 10))
plt.title('Iris Hierarchical Clustering Dendrogram')
plt.xlabel('Species')
dendrogram = sch.dendrogram(sch.linkage(iris.data[:,[0,1]], method='ward'))
```

```python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='single')
model.fit(iris.data[:,[0,1]])
labels = model.labels_
print(confusion_matrix(labels, iris.target))
```

```python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
model.fit(iris.data[:,[0,1]])
labels = model.labels_
print(confusion_matrix(labels, iris.target))
```

```python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='average')
model.fit(iris.data[:,[0,1]])
labels = model.labels_
print(confusion_matrix(labels, iris.target))
```

```python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
model.fit(iris.data[:,[0,1]])
labels = model.labels_
print(confusion_matrix(labels, iris.target))
```
